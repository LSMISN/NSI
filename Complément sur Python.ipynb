## Exercices : écrire une fonction palindrome qui retourne vrai si le texte passé en paramètre est un palindrome

```python
def palindrome(phrase):
    """
        renvoie True si la variable texte passée en paramètre
        est un palindrome
    """
    estPalindrome = True
    for i in range(0, len(phrase)):
        if phrase[i] != phrase[len(phrase)-1-i]:
            estPalindrome = False 
        # i ⁼ 0 phrase[0] == phrase[5-1-0] en supposant une
        # phrase de longueur 5
    return estPalindrome

print(palindrome("kayak")) # devrait renvoyer True
print(palindrome("mdr")) # devrait renvoyer False 
```

## exercice :

Écrivez une fonction qui détermine si une chaîne contient ou non un caractère passé en paramètre.

```python
def possedeCaractere(phrase, lettre):
    """
        Renvoie True lorsque la lettre est présente
        dans la phrase
    """
    # SI lettre est dans phrase
    if lettre.lower() in phrase.lower():
    #  ALORS 
        return True
    #    renvoyer VRAI
    # SINON
    else:
    #    renvoyer FALSE
        return False
print(possedeCaractere("La capitalE dE la FinlandE", "e")) # True
print(possedeCaractere("La capitale de la Finlande", "o")) # False
print("La CaPiTALE De LA FInLAndE".upper())
print("La CaPiTALE De LA FInLAndE".lower())
print("la capitale de la finlande".capitalize())
```

## exercice

Écrivez une fonction qui compte le nombre d’occurrences du 
caractère passé en paramètre dans une
chaîne.

```python
def compteOccurence(phrase, lettre):
    """
        renvoie le nombre d'occurence d'une lettre dans phrase
    """
    cpt = 0
    for i in range(len(phrase)):
        if phrase[i] == lettre:
            cpt = cpt + 1
    return cpt
    
```
